```python
df=pd.read_csv("GlobalWeatherRepository.csv")
```

```python
df.isnull().any(axis=1)
```

Handle INCONSISTENT entries

```python
# Negative rainfall
df[df['precip_mm']< 0]
```

```python
# Negative Wind Speed
df[df['wind_mph']<0]
```

```python
# Check Humidity Range (0–100% only)
df[(df['humidity'] < 0 )| (df['humidity'] > 100 )]
```

```python
# Check Date Format Consistency
df['last_updated']= pd.to_datetime(df['last_updated'],dayfirst=True,errors='coerce')
print(df['last_updated'])
```

```python
# check Numeric columns which contains text 
df['temperature_celsius']=pd.to_numeric(df['temperature_celsius'],errors='coerce')
print(df['temperature_celsius'])
```

```python
# Check Categorical Inconsistencies (Typos)
df['condition_text'].value_counts()
```

```python
df.duplicated().sum()
```
