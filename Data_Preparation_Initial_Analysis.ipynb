```python
#check the last_updated datavalues in the dataframe

df['last_updated'] = pd.to_datetime(df['last_updated']).dt.year
df['last_updated'].value_counts().sort_index()
```

```python
#Convert units and normalize the values (min-max normalization i.e values between 0 and 1)

num_cols = df.select_dtypes(include = 'number').columns
df_normalized = df.copy()
df_normalized[num_cols] = (df[num_cols] - df[num_cols].min()) / (df[num_cols].max() - df[num_cols].min())

df_normalized.head()
```

```python
#Aggregate or filter data (e.g., daily to monthly averages).
    
df['date'] = pd.to_datetime(df['last_updated_epoch'], unit='s')
df.set_index('date', inplace=True)
num_cols = df.select_dtypes(include='number').columns
monthly_avg = df[num_cols].resample('ME').mean()
monthly_avg


```
