```python
df[df["temperature_celsius"] < -80]    # unrealistic
df[df["temperature_celsius"] > 60]     # unrealistic
```

```python
df = df.dropna(subset=["temperature_celsius"])
```

```python
df["humidity"] = df["humidity"].fillna(df["humidity"].median())
df = df.fillna(df.median(numeric_only=True))
```

```python
df["last_updated"] = pd.to_datetime(df["last_updated"], errors="coerce")
df = df.dropna(subset=["last_updated"])
```

```python
df = df[(df["temperature_celsius"] > -80) & (df["temperature_celsius"] < 60)]
df = df[df["precip_mm"] >= 0]
df = df[df["wind_kph"] >= 0]
```

```python
df["year"] = df["last_updated"].dt.year
df["month"] = df["last_updated"].dt.month

monthly_df = df.groupby(["year", "month"], as_index=False).mean(numeric_only=True)
monthly_df.head()
```
