```python
import pandas as pd
import plotly.express as px

# Load cleaned dataset
df = pd.read_csv("Cleaned_Weather_Data.csv")

# Use temperature_celsius column directly
temp_col = "temperature_celsius"

# Group: Country-wise average temperature
country_avg = (
    df.groupby("country", as_index=False)[temp_col]
      .mean()
      .rename(columns={temp_col: "avg_temp_c"})
)

print(country_avg.head())  # check

# Plot choropleth map
fig = px.choropleth(
    country_avg,
    locations="country",               # use country names
    locationmode="country names",      # important!
    color="avg_temp_c",
    hover_name="country",
    color_continuous_scale="thermal",
    title="World Map - Average Temperature by Country (°C)",
    labels={"avg_temp_c": "Temperature (°C)"},
    projection="natural earth"
)

fig.update_layout(coloraxis_colorbar=dict(title="°C"))

fig.show()
```

```python
heat_df = df.groupby('country')[['temperature_celsius', 'humidity']].mean().reset_index()
```

```python
fig = px.density_heatmap(
    heat_df,
    x='country',
    y='temperature_celsius',
    z='humidity',
    title="Temperature vs Humidity Heatmap (Country-wise)",
    nbinsx=20,
    nbinsy=20,
    color_continuous_scale='Turbo'
)

fig.update_layout(xaxis_tickangle=-90)
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="humidity",
    y="temperature_celsius",
    title="Extreme Temperature Events (Hot > 40°C, Cold < 5°C)",
    labels={"humidity": "Humidity (%)", "temperature_celsius": "Temperature (°C)"},
    color=df['temperature_celsius'].apply(
        lambda x: "Hot (>40°C)" if x > 40 else ("Cold (<5°C)" if x < 5 else "Normal")
    ),
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="temperature_celsius",
    y="wind_kph",
    title="Extreme Wind Events (Wind > 50 km/h)",
    labels={"temperature_celsius": "Temperature (°C)", "wind_kph": "Wind Speed (km/h)"},
    color=df['wind_kph'] > 50
)

fig.show()
```

```python
# Identify extreme events
extreme_hot = df[df['temperature_celsius'] > 40]        # Heatwave
extreme_cold = df[df['temperature_celsius'] < 5]        # Coldwave
extreme_wind = df[df['wind_kph'] > 50]                  # High wind

print("Extreme Hot Count:", len(extreme_hot))
print("Extreme Cold Count:", len(extreme_cold))
print("Extreme High Wind Count:", len(extreme_wind))
```

```python
import pandas as pd
import plotly.express as px

extreme_counts = pd.DataFrame({
    "Event": ["Heatwave (>40°C)", "Coldwave (<5°C)", "High Wind (>50 km/h)"],
    "Count": [len(extreme_hot), len(extreme_cold), len(extreme_wind)]
})

fig = px.bar(
    extreme_counts,
    x="Event",
    y="Count",
    title="Extreme Weather Events Count",
    color="Event",
    text="Count"
)

fig.update_layout(xaxis_title="Event Type", yaxis_title="Count")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("Cleaned_Weather_Data.csv")

# ---------- Helper: Show title and figure ----------
def show_section(title, fig):
    print("\n" + "="*80)
    print(title)
    print("="*80)
    fig.show()


# 1. Temperature Distribution
fig = px.histogram(df, x="temperature_celsius", nbins=40,
                   title="Temperature Distribution")
show_section("1. Temperature Distribution", fig)


# 2. Humidity Distribution
fig = px.histogram(df, x="humidity", nbins=40,
                   title="Humidity Distribution")
show_section("2. Humidity Distribution", fig)


# 3. Wind Speed Distribution
fig = px.histogram(df, x="wind_kph", nbins=40,
                   title="Wind Speed Distribution")
show_section("3. Wind Speed Distribution", fig)


# 4. Pressure Distribution
fig = px.histogram(df, x="pressure_mb", nbins=40,
                   title="Pressure Distribution")
show_section("4. Pressure Distribution", fig)


# 5. Temperature vs Humidity
fig = px.scatter(df, x="temperature_celsius", y="humidity",
                 title="Temperature vs Humidity", opacity=0.7)
show_section("5. Temperature vs Humidity", fig)


# 6. Temperature vs Wind Speed
fig = px.scatter(df, x="temperature_celsius", y="wind_kph",
                 title="Temperature vs Wind Speed")
show_section("6. Temperature vs Wind Speed", fig)


# 7. Temperature vs Humidity vs Wind (3D)
fig = px.scatter_3d(df,
                    x="temperature_celsius",
                    y="humidity",
                    z="wind_kph",
                    title="Temperature vs Humidity vs Wind (3D)")
show_section("7. 3D Temperature–Humidity–Wind", fig)


# 8. Avg Temperature by Country
avg_temp = df.groupby("country")["temperature_celsius"].mean().reset_index()
fig = px.bar(avg_temp, x="country", y="temperature_celsius",
             title="Avg Temperature by Country")
show_section("8. Avg Temperature by Country", fig)


# 9. Avg Humidity by Country
avg_h = df.groupby("country")["humidity"].mean().reset_index()
fig = px.bar(avg_h, x="country", y="humidity",
             title="Avg Humidity by Country")
show_section("9. Avg Humidity by Country", fig)


# 10. Avg Wind by Country
avg_w = df.groupby("country")["wind_kph"].mean().reset_index()
fig = px.bar(avg_w, x="country", y="wind_kph",
             title="Avg Wind Speed by Country")
show_section("10. Avg Wind Speed by Country", fig)


# 11. Weather Condition Frequency
cond = df["condition_text"].value_counts().reset_index()
cond.columns = ["condition_text", "count"]
fig = px.bar(cond, x="condition_text", y="count",
             title="Weather Condition Frequency")
show_section("11. Weather Condition Frequency", fig)


# 12. Temperature Violin (Weather Condition)
fig = px.violin(df, x="condition_text", y="temperature_celsius",
                box=True, color="condition_text",
                title="Temperature Distribution by Condition (Violin Plot)")
show_section("12. Temperature Violin (Condition)", fig)


# 13. Correlation Heatmap
corr = df[["temperature_celsius", "humidity", "wind_kph", "pressure_mb"]].corr()
fig = px.imshow(corr, text_auto=True, title="Correlation Heatmap")
show_section("13. Correlation Heatmap", fig)


# 14. World Temperature Map
avg_temp_map = avg_temp.copy()
fig = px.choropleth(avg_temp_map, locations="country",
                     locationmode="country names",
                     color="temperature_celsius",
                     title="World Avg Temperature Map")
show_section("14. World Temperature Map", fig)


# 15. Top 10 Hottest Countries
hot = avg_temp.sort_values(by="temperature_celsius", ascending=False).head(10)
fig = px.bar(hot, x="country", y="temperature_celsius",
             title="Top 10 Hottest Countries")
show_section("15. Top 10 Hottest Countries", fig)


# 16. Top 10 Coldest Countries
cold = avg_temp.sort_values(by="temperature_celsius", ascending=True).head(10)
fig = px.bar(cold, x="country", y="temperature_celsius",
             title="Top 10 Coldest Countries")
show_section("16. Top 10 Coldest Countries", fig)


# 17. Humidity Outliers
fig = px.box(df, y="humidity", title="Humidity Outliers")
show_section("17. Humidity Outliers", fig)


# 18. Overall Temperature Violin
fig = px.violin(df, y="temperature_celsius", box=True,
                title="Overall Temperature Violin Plot")
show_section("18. Overall Temperature Violin Plot", fig)

            
```
