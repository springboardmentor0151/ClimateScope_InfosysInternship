```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
date_cols = [col for col in df.columns if "date" in col.lower()]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

df[date_cols].head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.head()
```

```python
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Example conversions:
# Fahrenheit → Celsius
if 'temperature' in df.columns:
    df['temperature_C'] = (df['temperature'] - 32) * (5/9)

# Wind speed mph → m/s
if 'wind_speed' in df.columns:
    df['wind_speed_ms'] = df['wind_speed'] * 0.44704
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
summary = {
    "Total Rows": df.shape[0],
    "Total Columns": df.shape[1],
    "Numeric Columns": list(numeric_cols),
    "Categorical Columns": list(categorical_cols),
    "Missing Value Columns": df.isnull().sum()[df.isnull().sum()>0].to_dict(),
}

summary
```
