```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("GlobalWeatherRepository.csv")
```

```python
df.columns
```

```python
# Unit Convertion

if 'temperature_fahrenheit' in df.columns:
    df['temperature_celsius']=(df['temperature_fahrenheit']-32)*5/9
    print("temperature in celsius ")
    print(df['temperature_celsius'])

if 'wind_mph' in df.columns:
    df['wind_kph']=df['wind_mph']*1.60934
    print("wind in kilo per hours")
    print(df['wind_kph'])

if 'pressure_in'in df.columns:
    df['pressure_mb']=df['pressure_in']*33.8639
    print("pressure in millibars ")
    print(df['pressure_mb'])

if 'visibility_miles'in df.columns:
    df['visibility_km']=df['visibility_miles']*1.60934
    print("visibility in kilo meters")
    print(df['visibility_km'])

if 'precip_in' in df.columns:
    df['precip_mm']=df['precip_in']*25.4
    print("precip in millimeters ")
    print(df['precip_mm'])
```

```python
# Normalization

cols_to_normalize=['temperature_celsius', 'humidity', 'wind_mph','pressure_mb','visibility_km']

cols_to_normalize=[col for col in cols_to_normalize if col in df.columns]

scaler=MinMaxScaler()

df[[col +"_Norm" for col in cols_to_normalize]]=scaler.fit_transform(df[cols_to_normalize])
```

```python
print("Unit Conversion And Normalization is Complete ")
print(df.head())
```

Aggregate or filter data (e.g., daily to monthly averages).


```python
# Convert date column to datetime
df['last_updated'] = pd.to_datetime(df['last_updated'], dayfirst=True, errors='coerce')

# Group by Year and Month â†’ calculate monthly averages for ALL numeric columns
monthly_avg = df.groupby(
    [df['last_updated'].dt.year, df['last_updated'].dt.month]
).mean(numeric_only=True)

# Rename index for better readability (optional)
monthly_avg.index.names = ['Year', 'Month']

print(monthly_avg)
```

```python
df.to_csv("GlobalWeatherCleaned.csv",index=False)
```
